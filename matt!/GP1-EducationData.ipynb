```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

%matplotlib inline
```

```python
csv=pd.read_csv('hsgraddata.csv')
csv
```

```python
csvhs=pd.DataFrame(csv)

csvhs=csvhs.drop('Unnamed: 1',1)

csvhs=csvhs.drop('Unnamed: 2', 1)


csvhs.columns= (['States', '2014-15 grad_rates', 'White', 'Black', 'Hispanic',
       'Asian Pacific Islander ','Asian Pacific Islander',
       'Students with disabil-ities', 'Limited English Proficient',
       'Economically disadvan-taged']
      )

csvhs = csvhs.drop(0)
top5hs= csvhs.sort_values('2014-15 grad_rates', ascending=False)
top5hs=top5hs.head(5)
top5hs
```

```python
csvhs["2014-15 grad_rates"].mean()
```

```python
def make_colors(grad_rate):
    if grad_rate < 80:
        return "red"
    return "aqua"
print(grad_rate)
```

```python
colors = sns.choose_diverging_palette()
```

```python
top_bot = csvhs.loc[(csvhs["2014-15 grad_rates"] < 77) | (csvhs["2014-15 grad_rates"] > 88)].copy()

top_bot["diff_col"] = csvhs["2014-15 grad_rates"] - csvhs["2014-15 grad_rates"].mean()
top_bot.sort_values(by="2014-15 grad_rates", ascending=False, inplace=True)

plt.figure(figsize=(20,10))

sns.barplot(x="States", y="diff_col",data=top_bot, palette=colors)
plt.ylim(-15,10)
# top_bot.plot.bar(x="States", y="diff_col")
plt.show()
```

```python
fig, ax = plt.subplots(1)

bottom = csvhs.loc[(csvhs["2014-15 grad_rates"] < 77)]
top = csvhs.loc[(csvhs["2014-15 grad_rates"] > 88)]

bottom.plot.bar(x="States", y="2014-15 grad_rates", ax=ax)
top.plot.bar(x="States", y="2014-15 grad_rates", ax=ax)

plt.show()
```

```python
x_axis= np.arange(len(csvhs['States']))

colors_array = csvhs["2014-15 grad_rates"].apply(make_colors).values

plt.figure(figsize=(60, 15))


# Create the ticks for our bar chart's x axis

tick_loc = [value for value in x_axis]
plt.xticks(tick_loc, csvhs['States'], rotation='vertical',fontsize=13)
plt.bar(x_axis, csvhs['2014-15 grad_rates'], color=colors_array, alpha= .5, align='edge', width=.7)
plt.tick_params(labelsize=40)
xlim =(0, len(x_axis))
ylim=(0,100)

plt.savefig('hsgraddata.png')
plt.show()
```
